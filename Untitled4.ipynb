```
import tensorflow as tf
from tensorflow import keras
from keras.utils import np_utils
import numpy as np

# 데이터셋 로딩

fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# 데이터셋 전처리

# 2차원을 1차원으로 형태를 바꿈
train_images = train_images.reshape(60000, 28*28)
test_images = test_images.reshape(10000, 28*28)

# 정수를 실수로 바꿈 0 ~ 255 > 0.0 ~ 255.0
train_images = train_images.astype('float32')
test_images = test_images.astype('float32')

# 정규화 0.0~255.0 >> 0.0~1.0
train_images = train_images / 255.0
test_images = test_images / 255.0

# 원핫인코딩
train_labels = np_utils.to_categorical(train_labels)
test_labels = np_utils.to_categorical(test_labels)

print(train_images.shape)
print(test_labels.shape)
print(train_images.shape)
print(test_labels.shape)
```

```
import matplotlib.pyplot as plt

pixels = train_images[0].reshape(28, 28)
plt.imshow(pixels, cmap='gray')
plt.show()
```

```
model = tf.keras.Sequential()

# 네트워크의 아키텍처를 구성하고,
model.add(tf.keras.layers.Dense(32, input_dim = 28*28, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))

# 컴파일 수행
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
# 학습
hist = model.fit(train_images, train_labels, batch_size=32, epochs=200, validation_split=0.3)
```

```
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```
